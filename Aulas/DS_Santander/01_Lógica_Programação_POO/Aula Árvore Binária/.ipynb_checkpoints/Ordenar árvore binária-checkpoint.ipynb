```python
grafos = [
    [1],
    [3],
    [1],
    [2, 4],
    [5],
    [7],
    [4],
    [6]
]

def bfs(grafo, vertice_fonte):
    visitados, fila = list(), [vertice_fonte]
    while fila: # len(fila) > 0
        vertice = fila.pop(0)
        if vertice not in visitados:
            visitados.append(vertice)
            fila.extend([x for x in grafo[vertice] if x not in visitados])
    return visitados

bfs(grafos, 0)
```
