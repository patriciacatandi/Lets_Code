**1)** Crie um gráfico de barras para os seguintes dados sobre os valores de cursos para algumas linguagens de programação, ordenando os valores do __maior__ para o __menor__:

```python
cursos = ['Python', 'Java', 'C++', 'C#', 'Spark', 'R']

valores = [310, 220, 180, 200, 400, 80]
```

Faça o gráfico customizado, colocando título, legendas, mudando cor da visualização e outros alterações que achar necessãrias.

### Resolução: 

Primeiramente vamos carregar as bibliotecas necessárias para o exercício:

````python
# import das Libs
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
# Dados sobres os nomes dos cursos
cursos = ['Python', 'Java', 'C++', 'C#', 'Spark', 'R']

# Valores respectivos dos cursos
valores = [310, 220, 180, 200, 400, 80]
````

````python
# Define o tamanho da figura
fig, ax = plt.subplots(figsize = (10,6))

#define um dataframe com os dados
df = pd.DataFrame(valores, index=cursos, columns=['valores'])

# # cria o gráfico de barras, passando um valor para color

df.sort_values(by='valores', ascending=False).plot.bar(ax=ax, color='blue')

# Colocando um titulo no gráfico
plt.title('Valores dos cursos na Lets Code', fontsize = 15)

# Colocando nome nos eixos agora
plt.xlabel('Cursos', fontsize = 15)
plt.ylabel('Valores', fontsize = 15)

# Ajustando os nomes de acordo com um certo ângulo
plt.xticks(rotation=45, fontsize = 12)
plt.yticks(fontsize = 12)

#Ajustando o Plot a tela
plt.tight_layout()

#Remove a legenda
ax.get_legend().remove()

#Mostra o nosso gráfico
plt.show()
````

```python

qntde_alunos = [120, 80, 100, 40, 60]

```

Crie um gráfico de setores (mais conhecido como gráfico de Pizza), para criar uma visualização desses dados e também crie um gráfico de barras ordenando do __maior__ valor para o __menor__ valor. Compare as duas visualizações.

### Resolução:

Primeiramente vamos carregar as bibliotecas necessárias para o exercício:

````python
# import das Libs
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
# Cria uma lista com os nomes dos cursos
turmas = ['Python PRO', 'DS', 'Full Stack', 'PI', 'Degree']

# Cria uma lista com quantidades equivalentes para os respectivos cursos
qntde_alunos = [120, 80, 100, 40, 60]
````

````python
# Definindo um subplto para os gráficos
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# transforma os dados em um dataframe
df = pd.DataFrame(data=qntde_alunos, index=turmas, columns=['qntde_alunos'])

# Crie o gráfico de setores
df.plot.pie(y='qntde_alunos', autopct='%.1f%%', ax=ax[0])

# Customizações do gráfico de setores
ax[0].get_legend().remove()
ax[0].set_title('Distribuição (%) de Alunos por Turma', fontsize=16)
ax[0].set_ylabel('')

#Ordenando os dados e criando um gráfico de barras
df.sort_values(by='qntde_alunos', ascending=False).plot.bar(ax=ax[1])

# Customização do gráfico de barras
ax[1].get_legend().remove()
plt.title('Quantidade de alunos por curso', fontsize=16)

# Coloca nome no eixo X
plt.xlabel('Curso', fontsize=14)

# Coloca nome no eixo y4
plt.ylabel('Quantidade', fontsize=14)

# Mostra o gráfico
plt.show()
````

### Resolução:

__A)__ Filtre os dados para trabalharmos apenas com o segundo trimestre de 2020;

Primeiramente, vamos carregar as bibliotecas que serão necessárias para o exercício:

````python
# Import das principais libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

Precisamos também carregar a base de dados `IBOV.csv`:

````python
# Carrega a base de dados
ibov = pd.read_csv('IBOV.csv')
````

````python
# Print das primeiras linhas da base de dados
ibov.head()
````

````python
# Filtro do segundo trimestre
# Dica.: Quando não souber o último dia de um mês especifico, utilize o primeiro dia do mês
#        seguinte e utilize < ao invés do <= para filtrar o dataset
ibov_2T =  ibov[(ibov['Date'] >= '2020-04-01') & (ibov['Date'] < '2020-07-01')]
````

````python
# Importante resetar o index para trabalharmos com uma cópia do DataFrmae e não no original
ibov_2T = ibov_2T.reset_index(drop=True)

# Print das primeiras linhas do novo dataframe
ibov_2T.head()
````

__B)__ Crie uma coluna chamada de `Amplitude`, onde esse dado será a diferença entre o valor da alta no dia (`High`) menos o valor da baixa no dia (`Low`);

````python
# Cria a coluna Amplitude, onde é a diferença entre a coluna High e Low
ibov_2T['Amplitude'] = ibov_2T['High'] - ibov_2T['Low']
````

````python
# Print das primeiras linhas do novo dataframe
ibov_2T.head()
````

__C)__ Crie um gráfico que demonstre a variação da `Amplitude` ao longo do segundo trimestre.

````python
# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Gráfico de Linha
plt.plot('Date', 'Amplitude', data = ibov_2T)

# Coloca o titulo
plt.title('Amplitude Diária - Bovespa', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Data', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Fechamento', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(rotation = 90, fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Mostra o nosso Gráfico
plt.show()
````

__D)__ Faça o mesmo gráfico para a `Amplitude` e adicionalmente coloque a linha de valor médio no período e marque todos os pontos que sejam considerados como _outliers_.

````python
# Calculo do quartil
ibov_2T.Amplitude.quantile(0.25)
````

````python
# Ajustando o tamanho da Figura
plt.figure(figsize = (16,8))

# Gráfico de Linha
plt.plot('Date', 'Amplitude', data = ibov_2T)

#adiciona uma linha horizontal com a média do período
plt.axhline(ibov_2T.Amplitude.mean(), ls=':', label='Amplitude média')

#adiciona uma linha horizontal com a média do período
Q1 = ibov_2T.Amplitude.quantile(0.25)
Q3 = ibov_2T.Amplitude.quantile(0.75)
IQR = Q3 - Q1

# Definição de quem são os outliers
outliers = ibov_2T[(ibov_2T.Amplitude < Q1 - IQR*1.5) | (ibov_2T.Amplitude > Q3 + IQR*1.5)]
plt.scatter(outliers.Date, outliers.Amplitude, ls=':', label='Outliers', color='red')

# Coloca o titulo
plt.title('Amplitude Diária - Bovespa', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Data', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Variação', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(rotation = 90, fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

plt.legend()
# Mostra o nosso Gráfico
plt.show()
````

### Resolução:

__A)__ Filtre os dados para trabalharmos apenas com o terceiro trimestre de 2020;

Primeiramente, vamos carregar as bibliotecas que serão necessárias para o exercício:

````python
# Import das principais libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

Precisamos também carregar a base de dados `IBOV.csv`:

````python
# Carrega a base de dados
ibov = pd.read_csv('IBOV.csv')
````

````python
# Print das primeiras linhas da base de dados
ibov.head()
````

````python
# Filtro do terceiro trimestre
# Dica.: Quando não souber o último dia de um mês especifico, utilize o primeiro dia do mês
#        seguinte e utilize < ao invés do <= para filtrar o dataset
ibov_3T =  ibov[(ibov['Date'] >= '2020-07-01') & (ibov['Date'] < '2020-10-01')]
````

````python
# Print das primeiras linhas do novo dataframe
ibov_3T.head()
````

__B)__ Crie uma visualização que vai ser o conjunto de 3 gráficos:

   - Um gráfico de linha para representar o valor de fechamento das ações da Bovespa no período;
   - Dois gráficos de dispersão (Scatter) para identificar em cada um dos dias qual foi o valor de máxima e mínima da ação.
   
__Dica.:__ Crie todos os 4 gráficos na mesma célula do notebook, para que todos sejam criados na mesma visualização!

````python
# Ajustando o tamanho da figura
plt.figure(figsize = (15, 10))

# Criando um gráfico de dispersão para as máximas
plt.scatter('Date', 'High', data = ibov_3T)

# Criando um gráfico de dispersão para as minimas
plt.scatter('Date', 'Low', data = ibov_3T)

# Criano um gráfico de linha para o valor de fechamento
plt.plot('Date', 'Close', data = ibov_3T)

# Coloca o titulo
plt.title('Bovespa - 3 Trimestre', fontsize = 15)

# Titulo para eixo X
plt.xlabel('Data', fontsize = 15)

# Titulo para eixo Y
plt.ylabel('Fechamento', fontsize = 15)

# Tamanho da fonte da grade X
plt.xticks(rotation = 90, fontsize = 12)

# Tamanho da fonte da grade Y
plt.yticks(fontsize = 12)

# Legenda do Gráfico
plt.legend(["Fechamento", "Max.", "Min"]) # adicionamos uma legenda para facilitar para nosso leitor

# Mostra o gráfico
plt.show()
````

__C)__ Crie um função que receba um _DataFrame_ do _Pandas_ e crie os gráficos acima;

````python
# Definição de uma função que recebe o dataframe e o titulo e cria o gráfico
def cria_grafico(df, title):
    # Ajustando o tamanho da figura
    plt.figure(figsize = (15, 10))

    # Criando um gráfico de dispersão para as máximas
    plt.scatter('Date', 'High', data = df)

    # Criando um gráfico de dispersão para as minimas
    plt.scatter('Date', 'Low', data = df)

    # Criano um gráfico de linha para o valor de fechamento
    plt.plot('Date', 'Close', data = df)

    # Coloca o titulo
    plt.title(title, fontsize = 15)

    # Titulo para eixo X
    plt.xlabel('Data', fontsize = 15)

    # Titulo para eixo Y
    plt.ylabel('Fechamento', fontsize = 15)

    # Tamanho da fonte da grade X
    plt.xticks(rotation = 90, fontsize = 12)

    # Tamanho da fonte da grade Y
    plt.yticks(fontsize = 12)

    # Legenda do Gráfico
    plt.legend(["Fechamento", "Max.", "Min"]) # adicionamos uma legenda para facilitar para nosso leitor

    # Mostra o gráfico
    plt.show()
````

__D)__ Ytilizando a função, crie os gráficos para o primeiro e segundo trimestre dos dados.

````python
# Filtro do primeiro trimestre
# Dica.: Quando não souber o último dia de um mês especifico, utilize o primeiro dia do mês
#        seguinte e utilize < ao invés do <= para filtrar o dataset
ibov_1T =  ibov[(ibov['Date'] >= '2020-01-01') & (ibov['Date'] < '2020-04-01')]
````

````python
# Chama a função e cria o gráfico
cria_grafico(ibov_1T, 'Bovespa - 1 trimestre')
````

````python
# Filtro do segundo trimestre
# Dica.: Quando não souber o último dia de um mês especifico, utilize o primeiro dia do mês
#        seguinte e utilize < ao invés do <= para filtrar o dataset
ibov_2T =  ibov[(ibov['Date'] >= '2020-04-01') & (ibov['Date'] < '2020-07-01')]
````

````python
cria_grafico(ibov_2T, 'Bovespa - 2 trimestre')
````
