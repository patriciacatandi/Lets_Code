### Questão 2

Uma empresa de tecnologia solicitou que você cria-se um banco de dados para armazenar as informações dos funcionários, onde os dados que temos a respeito deles são:

- ID do Funcionário;
- Nome do Funcionário;
- Departamento
- Senioridade;
- Salário.

Portanto, utilizando a conexão definida no exercício anterior, crie a tabela `funcionarios` com as especificações para os dados acima.

### Resolução:

Primeiramente, vamos carregar as bibliotecas necessárias:

````python
# Import das bibliotecas necessárias
import mysql.connector
import pandas as pd
````

Vamos então passar todas as informações que precisamos para definir a conexão com o banco de dados:

````python
# Dados cadastrados no DB4Free
# OBS.: Aqui vocês precisam só alterar para os dados de vocês (o cadastro do exemplo também funciona)
nome_do_usuario = 'ssaorin'
senha = 'letscode'
nome_DB = 'saorin_database'
````

````python
# Dados para a conexão ao DB4Free
user = nome_do_usuario
password = senha
host = 'db4free.net'
database = nome_DB
port = 3306
````

Por fim, definindo a conexão e criando a tabela `funcionarios`:

````python
# Conectando no Banco de Dados
con = mysql.connector.connect(user=user, 
                              password=password,
                              host=host,
                              database=database, port=port)

# Escrevendo a Query em SQL
sql = '''
      CREATE TABLE funcionarios(Funcionario_ID INT,
                                Nome           CHAR(255),
                                Departamento   CHAR(255),
                                Senioridade    CHAR(255),
                                Salario        INT)
      '''

# Iniciando o conector
cursor = con.cursor(buffered=True)

# Deleta a tabela se ela existir anteriormente
cursor.execute('DROP TABLE IF EXISTS funcionarios')

# Roda a query definida anteriormente
cursor.execute(sql)

# Fecha o cursor
cursor.close()

# Fecha a conexão
con.close()
````

### Resolução:

__A)__ Ordene a lista de Funcionários por Departamento e Senioridade;

````python
# Define a Query
query = '''
        SELECT
            *
        FROM funcionarios
        ORDER BY Departamento, Senioridade
        '''

# Define a conexão
con = mysql.connector.connect(user=user,
                              password=password,
                              host=host,
                              database=database)

# Inicializa o cursor
cursor = con.cursor()

#usando o read_sql_query
consulta = pd.read_sql_query(query, con)

# Fecha o cursor
cursor.close()

# Fecha a conexão
con.close()

# Print do DataFrame da Consulta
consulta
````

__B)__ Calcule a quantidade de funcionários quebrado por Departamento e Senioridade;

__Dica:__ Use a função COUNT() do SQL.

````python
# Define a Query
query = '''
        SELECT
            Departamento,
            Senioridade,
            COUNT(Funcionario_ID) AS Quantidade
        FROM funcionarios
        GROUP BY 1, 2
        ORDER BY 1, 2
        '''

# Define a conexão
con = mysql.connector.connect(user=user,
                              password=password,
                              host=host,
                              database=database)

# Inicializa o cursor
cursor = con.cursor()

#usando o read_sql_query
consulta = pd.read_sql_query(query, con, index_col=['Departamento', 'Senioridade']).unstack()

# Fecha o cursor
cursor.close()

# Fecha a conexão
con.close()

# Print do DataFrame da Consulta
consulta
````

__C)__ Calcule a soma dos salários por Departamento;

__Dica:__ Use a função SUM() do SQL.

````python
# Define a Query
query = '''
        SELECT
            Departamento,
            SUM(Salario) AS total_salarios
        FROM funcionarios
        GROUP BY 1
        '''

# Define a conexão
con = mysql.connector.connect(user=user,
                              password=password,
                              host=host,
                              database=database)

# Inicializa o cursor
cursor = con.cursor()

#usando o read_sql_query
consulta = pd.read_sql_query(query, con)

# Fecha o cursor
cursor.close()

# Fecha a conexão
con.close()

# Print do DataFrame da Consulta
consulta
````

__D)__ Calcule a média salarial dos funcionários por Departamento e Senioridade;

__Dica:__ Use a função AVG() do SQL.

````python
# Define a Query
query = '''
        SELECT
            Departamento,
            Senioridade,
            AVG(Salario) AS media_salarios
        FROM funcionarios
        GROUP BY 1, 2
        ORDER BY 1, 2
        '''

# Define a conexão
con = mysql.connector.connect(user=user,
                              password=password,
                              host=host,
                              database=database)

# Inicializa o cursor
cursor = con.cursor()

#usando o read_sql_query
consulta = pd.read_sql_query(query, con, index_col=['Departamento', 'Senioridade']).unstack()

# Fecha o cursor
cursor.close()

# Fecha a conexão
con.close()

# Print do DataFrame da Consulta
consulta
````

__E)__ Calcule a média salarial dos funcionários por departamento, apenas para os que ganham menos que 8000 reais.

````python
# Define a Query
query = '''
        SELECT
            Departamento,
            AVG(Salario) AS media_salarios
        FROM funcionarios
        GROUP BY 1
        HAVING AVG(Salario) < 8000
        '''

# Define a conexão
con = mysql.connector.connect(user=user,
                              password=password,
                              host=host,
                              database=database)

# Inicializa o cursor
cursor = con.cursor()

#usando o read_sql_query
consulta = pd.read_sql_query(query, con)

# Fecha o cursor
cursor.close()

# Fecha a conexão
con.close()

# Print do DataFrame da Consulta
consulta
````
