### Oktoberfest
Oktoberfest é o maior festival de cerveja e é organizado em Munique desde 2018.
Dura entre 16 e 18 dias, do meio ou final de setembro ate o primeiro domingo de outubro com mais de 6 milhões de visitantes todo ano

Vamos pegar os dados do portal oficial da cidade de Munique :
[Clique aqui](https://www.opengov-muenchen.de/tr/dataset) 

Lembrando que :


    jahr : Ano
    dauer : Duracao
    besucher_gesamt (in Millionen) : Total de Visitantes (10^6)
    besucher_tag (in 1000) : Visitantes por dia (10³)
    bier_preis (1 Liter) : Preço cerveja (Por litro)
    bier_konsum (in Hektoliter) : Consumo de Cerveja em Hectolitros
    hendl_preis : Preço do Frango
    hendl_konsum : Consumo de Frango


```python
# importando a biblioteca

import pandas as pd
```

**Exercício 1:** Leia a base e informe o número de linhas e colunas

```python
# carregar o arquivo oktoberfest

caminho = 'C:/Users/Renata/Desktop/' # ou C:\\Users\\Renata\\Desktop\\ 
arquivo = 'oktoberfest.csv'

df =  pd.read_csv(caminho + arquivo) #  sep=',' -> o padrão é separado por vírgula

# para mais detalhes : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
```

```python
df.head(10) # mostrar as 10 primeiras linhas
```

```python
df.shape # mostra o número de linhas e colunas -> 34 linhas e 8 colunas
```

**Exercício 2:** Renomeie as colunas conforme

    jahr : Ano
    dauer : Duracao
    besucher_gesamt (in Millionen) : Total de Visitantes (10^6)
    besucher_tag (in 1000) : Visitantes por dia (10³)
    bier_preis (1 Liter) : Preço cerveja (Por litro)
    bier_konsum (in Hektoliter) : Consumo de Cerveja em Hectolitros
    hendl_preis : Preço do Frango
    hendl_konsum : Consumo de Frango

```python
df.rename(
    columns={
        'jahr': 'Ano',
        'dauer': 'Duracao',
        'besucher_gesamt': 'Total de Visitantes',
        'besucher_tag': 'Visitantes por dia',
        'bier_preis': 'Preço cerveja',
        'bier_konsum': 'Consumo de Cerveja',
        'hendl_preis': 'Preço do Frango',
        'hendl_konsum': 'Consumo de Frango'
    }, inplace=True)
```

**Exercício 3:** Salve em formato csv as colunas `ano` e `preco_cerveja` com separador `|` e salve em formato `xlsx`.

```python
# selecionar apenas as colunas ano e preço cerveja
df[['Ano', 'Preço cerveja']]
```

```python
# salva em csv

nome_do_arquivo = 'oktoberfest_v1.csv'
df[['Ano', 'Preço cerveja']].to_csv(caminho + nome_do_arquivo, index=False, sep='|')
```

```python
# salvar em xlsx

nome_do_arquivo = 'oktoberfest_v2.xlsx'
df[['Ano', 'Preço cerveja']].to_excel(caminho + nome_do_arquivo, index=False)
```

**Exercício 5:** Qual o total de cerveja consumido por ano?

```python
df.groupby('Ano')['Consumo de Cerveja'].sum()
```

**Exercício 6:** Qual o total de frango consumido por ano?

```python
df.groupby('Ano')['Consumo de Frango'].sum()
```

**Exercício 7:** Crie uma coluna **receita_cerveja** que representa o produto entre o preco e o consumo de cerveja por ano

```python
df['receita_cerveja'] = df['Preço cerveja'] * df['Consumo de Cerveja']
```

**Exercício 8:** Crie uma coluna **receita_frango** que representa o produto entre o preco e o consumo de frango por ano

```python
df['receita_frango'] = df['Preço do Frango'] * df['Consumo de Frango']
```

**Exercício 9:** Crie uma coluna **receita_total** que representa a receita total por ano do evento

```python
df['receita_total'] = df['receita_cerveja'] + df['receita_frango']
```

**Exercício 10:** Qual a média de cerveja consumida por ano?

```python
# agrupar por ano e tirar a média do consumo da cerveja
df.groupby('Ano')['Consumo de Cerveja'].mean()
```

**Exercício 11:** Qual é a média da receita total?

```python
df['receita_total'].mean()
```

Bônus: Qual foi a maior receita total?

```python
df['receita_total'].max()
```

- Gráfico de barras

```python
#*** gerando dados para poder desenhar *******
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars)) # arrya de índice


#*** desenhando *******

plt.bar(y_pos, height) # criar o gráfico de barras na 

# trocar valores dos marcadores do eixo x pelos rótulos A, B,C, D, E
plt.xticks(y_pos, bars)

plt.show()
```

- Gráfico de barras horizontal

```python
#*** gerando dados para poder desenhar *******
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars)) # arrya de índice


#*** desenhando *******

plt.barh(y_pos, height) # criar o gráfico de barras na 

# trocar valores dos marcadores do eixo x pelos rótulos A, B,C, D, E
plt.yticks(y_pos, bars)

plt.show()
```

- Gráfico de barras empilhado

```python
#*** gerando dados para poder desenhar *******
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35     

#*** desenhando *******
fig, ax = plt.subplots()


ax.bar(labels, men_means, width, yerr=men_std, label='Homens')
ax.bar(labels, women_means, width, yerr=women_std, label='Mulheres', bottom=men_means)

# plt.grid()

ax.set_title('Pontuação') #título
ax.legend()
ax.set_ylabel('Pontuação')
```

- Gráfico de pizza

```python
#*** gerando dados para poder desenhar *******
names = ['Grupo 1', 'Grupo 2', 'Grupo 3', 'Grupo 4']
values = [12, 11, 3, 30]
explode= (0, 0.1, 0, 0) # fatia

#*** desenhando *******
plt.pie(values, labels=names, explode=explode, shadow=True, startangle=30, autopct='%1.1f%%')

plt.axis('equal')

plt.show()
```

- Gráfico - Histograma

```python
# criando os dados (distribuição)
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# histograma
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') # adiciona o texto. caracter especial
plt.xlim(40, 160) # ajusta limites dos eixos
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```

- Gráfico Boxplot

## Personalizar/Customizar os gráficos

```python
# Criar alguns dados
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]


# personalizar o gráfico
plt.figure(figsize=(16,8)) #mudando o tamanho da figura (em polegadas)

plt.plot(x, y, color='green', linestyle='-.', linewidth=3, label='Nome da legenda') 
# trocamos a cor, o estilo da linha, aumentei a largura da linha, e dei nome a linha

plt.title('O Título do gráfico', color='red') # adicionando um título

plt.legend(loc='lower right') #adicionar a legenda e colocar no canto inferior direito

# nome para os eixos
plt.xlabel('Nome do eixo x', color='purple')
plt.ylabel('Nome do eixo y', color='b')

# Mudar a cor e a rotação dos nomes da escala
plt.xticks(color='orange', rotation=90)
plt.yticks(color='m', rotation=45)

# adicionar uma linha vertical
plt.axvline(2.4)
plt.axhline(11, color='cyan')

plt.grid() # inserir grades

plt.show()
```
